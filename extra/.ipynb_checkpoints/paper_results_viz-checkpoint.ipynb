```python
def _load_results(resource, sl_method, dl_method):
    file  = '{}-{}-{}.csv'.format(resource, sl_method, dl_method)
    test_ = pd.read_csv(path_to_results + r'{}_val_and_test/'.format(resource) + 'test-' + file)
    val_  = pd.read_csv(path_to_results + r'{}_val_and_test/'.format(resource) + 'val-' + file)
    
    test_ = test_.rename(columns = {'ES_': 'ES', 'IS_': 'IS', 'VS_': 'VS', 'sample_IS_': 'sample_IS'})
    val_  = val_.rename(columns = {'ES_': 'ES', 'IS_': 'IS', 'VS_': 'VS', 'sample_IS_': 'sample_IS'})

    print(file)
    return test_, val_

def _agg_scores(df_, scores_, key):
    df_[key] = 0.
    for score in scores_:
        df_[key] += df_[score]/len(scores_)
    return df_
    
def _load_zone_independet_results(sl_methods_, scores_, resource, dl_method):
    
    def __independent_model_selection(df_test_, df_val_, scores_):

        errors_ = []
        params_ = []
        for score in scores_:
            val_  = df_val_.loc[np.argmin(df_val_[score])]
            test_ = df_test_.loc[df_test_['experiment'] == val_['experiment']]
            errors_.append(test_[score].to_numpy()[0])
            params_.append(test_['parameters'].to_numpy()[0])

        return errors_, params_

    all_Scores_ = []
    for sl_method in sl_methods_:

        test_, val_ = _load_results(resource, sl_method, dl_method)
        
        Scores_ = []
        for score in scores_:
            columns_ = list(test_.columns.values[test_.columns.str.contains(score + '_')])

            values_, params_ = __independent_model_selection(test_, val_, columns_)

            Score_                    = pd.DataFrame(columns = ['sparse_learning', 'zone', 'score', 'value', 'parameters'])
            Score_['zone']            = columns_
            Score_['zone']            = Score_['zone'].apply(lambda x: x[-4:])
            Score_['score']           = score
            Score_['value']           = values_
            Score_['parameters']      = params_
            Score_['sparse_learning'] = sl_method
            Score_                    = Score_.reset_index(drop = True)
            Scores_.append(Score_)
            
        all_Scores_.append(pd.concat(Scores_, axis = 0))
        
    return pd.concat(all_Scores_, axis = 0).reset_index(drop = True)

Scores_ = []
for i_resource in [0, 1, 2]:
    for i_dl_method in [0, 1, 2]:

        Score_ = _load_zone_independet_results(sl_methods_, scores_   = ['CRPS', 'RMSE', 'MBE'], 
                                                            resource  = resources_[i_resource], 
                                                            dl_method = dl_methods_[i_dl_method])

        Score_['dense_learning'] = dl_methods_[i_dl_method]
        Score_['resource']       = resources_[i_resource]
        Scores_.append(Score_)
        
iScores_ = pd.concat(Scores_, axis = 0).reset_index(drop = True)
iScores_.to_csv(path_to_models + 'zone_score_independet_selection.csv', index = False)
print(iScores_)
```

```python
def _joint_model_selection(df_test_, df_val_, score, scoring_rule):
    val_  = df_val_.loc[np.argmin(df_val_[scoring_rule])]
    test_ = df_test_[df_test_['experiment'] == val_['experiment']]
    return test_[score].to_numpy()[0], test_['parameters'].to_numpy()[0]


def _load_joint_results(sl_methods_, scores_, resource, dl_method, scoring_rule):
        
    Scores_ = []

    for sl_method in sl_methods_:

        test_, val_ = _load_results(resource, sl_method, dl_method)
        
        for score in ['RMSE', 'MBE', 'CRPS']:

            columns_ = list(test_.columns.values[test_.columns.str.contains(score + '_')])

            test_ = _agg_scores(test_, columns_, score)
            val_  = _agg_scores(val_, columns_, score)
    
        values_ = [sl_method]
        for score in scores_:
            
            value, params = _joint_model_selection(test_, val_, score, scoring_rule)

            values_.append(value)
        Score_ = pd.DataFrame([params] + values_, index = ['parameters', 'sparse_learning'] + scores_).T

        Scores_.append(Score_)
        
    return pd.concat(Scores_, axis = 0)


Scores_ = []
for i_resource in [0, 1, 2]:
    for i_dl_method in [0, 1, 2]:
        #print(resources_[i_resource], dl_methods_[i_dl_method])

        Score_ = _load_joint_results(sl_methods_, scores_      = ['ES', 'VS', 'IS', 'RMSE', 'MBE'], 
                                                  resource     = resources_[i_resource], 
                                                  dl_method    = dl_methods_[i_dl_method], 
                                                  scoring_rule = 'IS')
    
        Score_['dense_learning'] = dl_methods_[i_dl_method]
        Score_['resource']       = resources_[i_resource]
        
        Scores_.append(Score_)

jScores_ = pd.concat(Scores_, axis = 0).reset_index(drop = True)
jScores_.to_csv(path_to_models + 'agg_selection.csv', index = False)
print(jScores_)
```

```python
def _load_joint_results(sl_methods_, scores_, resource, dl_method):
        
    Scores_ = []

    for sl_method in sl_methods_:

        test_, val_ = _load_results(resource, sl_method, dl_method)
        
        for score in ['RMSE', 'MBE', 'CRPS']:

            columns_ = list(test_.columns.values[test_.columns.str.contains(score + '_')])

            test_ = _agg_scores(test_, columns_, score)
            val_  = _agg_scores(val_, columns_, score)
    
        values_ = []
        params_ = []
        for score in scores_:
            
            value, params = _joint_model_selection(test_, val_, score, score)

            values_.append(value)
            params_.append(params)
            
        Score_ = pd.DataFrame(columns = ['sparse_learning', 'value', 'score', 'parameters'])
        Score_['value']           = values_
        Score_['parameters']      = params_
        Score_['score']           = scores_
        Score_['sparse_learning'] = sl_method
        Scores_.append(Score_)
        
    return pd.concat(Scores_, axis = 0)


Scores_ = []
for i_resource in [0, 1, 2]:
    for i_dl_method in [0, 1, 2]:
        #print(resources_[i_resource], dl_methods_[i_dl_method])

        Score_ = _load_joint_results(sl_methods_, scores_   = ['ES', 'VS', 'IS', 'RMSE', 'MBE'], 
                                                  resource  = resources_[i_resource], 
                                                  dl_method = dl_methods_[i_dl_method])
    
        Score_['dense_learning'] = dl_methods_[i_dl_method]
        Score_['resource']       = resources_[i_resource]
        
        Scores_.append(Score_)

sScores_ = pd.concat(Scores_, axis = 0).reset_index(drop = True)
sScores_.to_csv(path_to_models + 'score_independet_selection.csv', index = False)
print(sScores_)
```

```python
def _make_ref_plot(Scores_, zones_p_, score, resource, legend_labels_, legend_title, ndecimals = 0,
                                                                                     score_key = 'value',
                                                                                     ref_      = [],
                                                                                     xlabel    = '',
                                                                                     ylabel    = '', 
                                                                                     title     = '', 
                                                                                     suptitle  = '', 
                                                                                     letter    = '',
                                                                                     y_lim     = 0,
                                                                                     save      = False, 
                                                                                     legend    = False, 
                                                                                     filename  = 'noname.pdf'):
    
    methods_ = Scores_['method'].unique()
    zones_   = Scores_.loc[Scores_['resource'] == resource, 'zone'].unique()

    x_     = np.linspace(0, .333*len(methods_), len(methods_))
    width  = 0.4
    offset = 2.
    
    fig = plt.figure(figsize = (len(zones_)*2.5, 3.))
    ax  = plt.subplot(111)
    
    plt.suptitle(suptitle, fontsize = 20, 
                           y        = 1.075)
    
    plt.title(title, fontsize = 18)

    refs_ = [1000., 500., 750.]
    ticks_ = []
    j = 0
    for zone in zones_:
        i = 0
        for method in methods_:
            
            idx_  = (Scores_['method'] == method) & (Scores_['zone'] == zone) & (Scores_['resource'] == resource)
            value = Scores_.loc[idx_, score_key].to_numpy()[0]
            
            if len(ref_) > 0:
                y = np.around((1. - value/ref_[j])*100, ndecimals)
                p = plt.bar(x_[i], y, width, color  = colors_[i], 
                                             zorder = 9, 
                                             align  = 'center')
            else:
                y = np.around(value, ndecimals)
                p = plt.bar(x_[i], y, width, color  = colors_[i], 
                                             zorder = 9, 
                                             align  = 'center')
            plt.bar_label(p, labels=[f'{x:,.0f}' for x in p.datavalues], size = 14)
           
            i += 1
        
        ticks_.append(x_.mean())
        x_ += offset
        j += 1
    idx_  = Scores_['resource'] == resource
    if y_lim == 0.:
        y_lim = Scores_.loc[idx_, score].max()
    plt.ylim(-y_lim*0.005, y_lim*1.125)

    plt.xticks(ticks_[:len(zones_p_)], zones_p_, fontsize = 10)
    
    legend_ = []
    for i in range(len(legend_labels_)):
        legend_.append(Line2D([0], [0], marker          = 's', 
                                        markerfacecolor = colors_[i], 
                                        markeredgecolor = colors_[i], 
                                        markersize      = 6,
                                        label           = legend_labels_[i],
                                        color           = "None"))
    if legend:
        plt.legend(handles        = legend_, 
                   title          = legend_title, 
                   prop           = {'size': 14}, 
                   title_fontsize = 16,
                   ncol           = 1, 
                   columnspacing  = 0., 
                   handletextpad  = 0.,  
                   bbox_to_anchor = [1.4, .825],
                   frameon        = False)

    
    plt.grid(axis = 'y')

    ax.xaxis.set_tick_params(labelsize = 14, left = False)
    ax.yaxis.set_tick_params(labelsize = 14, left = False)
    
    ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

    ax.spines[['right', 'top', 'left', 'bottom']].set_visible(False)
    
    plt.ylabel(ylabel, fontsize = 16)
    plt.xlabel(xlabel, fontsize = 16)
    
    ax.axhline(0, linewidth = .5, 
                      linestyle = '-', 
                      color     = 'k', 
                      clip_on   = False, 
                      zorder    = 10)

    ax.text(-1.3, y_lim*1.25, letter, fontsize = 20, 
                                      weight   = "bold")
    
    
    if save:
        plt.savefig(path_to_images + filename, bbox_inches = 'tight',
                                               dpi         = 350)
    
    plt.show()
```

```python
# Rolling average window filter
def _smooth_average(X_, span): 
    N_dim, N_tasks, N_samples = X_.shape
    Y_                        = np.zeros(X_.shape)
    for i in range(N_dim):
        for j in range(N_samples):
            y_ = []
            for t in range(N_tasks):
                if t - span < 0: x = 0
                else:            x = t - span
                if t + span + 1 > N_tasks: y = N_tasks
                else:                      y = t + span + 1
                y_.append(np.average(X_[i, x:y, j]))
            Y_[i, :, j] = np.array(y_)
    return Y_

# SavitzkyGolay filter
def _smooth_savgol(X_, span, degree):  
    N_dim, N_tasks, N_samples = X_.shape
    Y_                        = np.zeros(X_.shape)
    for i in range(N_dim):
        for j in range(N_samples):
            Y_[i, :, j] = savgol_filter(X_[i, :, j], span * 2 + 1, degree)
    return Y_

# A LOWESS (Locally Weighted Scatterplot Smoothing) function that outs smoothed estimates 
def _smooth_lowess(X_, span):
    N_dim, N_tasks, N_samples = X_.shape
    Y_                        = np.zeros(X_.shape)
    x_                        = np.linspace(0, 1, N_tasks)
    for i in range(N_dim):
        for j in range(N_samples):
            Y_[i, :, j] = sm.nonparametric.lowess(X_[i, :, j], x_, frac          = 1/span, 
                                                                   return_sorted = False)
    return Y_

# def smooth_fft(X_, span, degree):  
#     N_dim, N_tasks, N_samples = X_.shape
#     Y_                        = np.zeros(X_.shape)
#     for i in range(N_dim):
#         for j in range(N_samples):
#             w_             = fftpack.rfft(X_[i, :, j])
#             spectrum       = w_ ** 2
#             cutoff_idx     = spectrum < (spectrum.max() * (1 - np.exp(-span / degree**2)))
#             w_[cutoff_idx] = 0
#             Y_[i, :, j]    = fftpack.irfft(w_)
#     return Y_

# Gaussian kernel filtering
def _smooth_gaussian(X_, sigma):
    N_dim, N_tasks, N_samples = X_.shape
    Y_                        = np.zeros(X_.shape)
    for i in range(N_dim):
        for j in range(N_samples):
            Y_[i, :, j] = gaussian_filter1d(X_[i, :, j], sigma)
    return Y_

# Evaluate a scenario's predictive probability
def _eval_proba(_model, Y_scens_, i_sel_sample):
    mu_      = _model['mean'][i_sel_sample, ...]
    Sigma_   = _model['covarice'][i_sel_sample, ...]

    Z_ = np.zeros((Y_scens_.shape[1], Y_scens_.shape[2]))
    for t in range(mu_.shape[1]):    
        Z_[t, :] = multivariate_normal(mu_[:, t], Sigma_[..., t]).pdf(Y_scens_[:, t, :].T)
    z_  = np.prod(Z_, axis = 0)
    z_ /= z_.sum()
    return z_

# Variogram Score computed across temporal dimensions
def _VS_temporal(Y_, Y_hat_, p = .5):
    N_observation, N_tasks, N_horizons, N_samples = Y_hat_.shape
    score_ = np.zeros((N_tasks, N_observation))
    for t in range(N_tasks):
        for n in range(N_observation):
            frac1_ = np.absolute(np.subtract.outer(Y_[n, t, :], Y_[n, t, :]))**p
            frac2_ = np.zeros((N_horizons, N_horizons))
            for m in range(N_samples):
                frac2_ += np.absolute(np.subtract.outer(Y_hat_[n, t, :, m], Y_hat_[n, t, :, m]))**p
            score_[t, n] = np.sum((frac1_ - (frac2_/N_samples))**2)
    return score_

# Energy Score
def _ES_temporal(Y_, Y_hat_):
    N_observation, N_tasks, N_horizons, N_samples = Y_hat_.shape
    score_ = np.zeros((N_tasks, N_observation))
    for t in range(N_tasks):
        for n in range(N_observation):
            Y_p_   = np.tile(Y_[n, t, :], (N_samples, 1)).T
            frac1_ = np.sqrt(np.diag((Y_hat_[n, t, :, :] - Y_p_).T @ (Y_hat_[n, t, :, :] - Y_p_)))
            frac2_ = 0
            for m in range(N_samples):
                frac2_ += np.sqrt(np.sum((Y_hat_[n, t, :, :].T - Y_hat_[n, t, :, m])**2, axis = 1))
            score_[t, n] = np.sum(frac1_)/N_samples - np.sum(frac2_)/(2*(N_samples**2))
    return score_

# Average variograme score
def _VS(Y_, Y_hat_):
    VS_ = _VS_temporal(Y_, Y_hat_, p = .5)
    VS_ = np.mean(VS_, axis = 1)
    VS_ = np.mean(VS_, axis = 0)
    return VS_

# Average energy score
def _ES(Y_, Y_hat_):
    ES_ = _ES_temporal(Y_, Y_hat_)
    ES_ = np.mean(ES_, axis = 1)
    ES_ = np.mean(ES_, axis = 0)
    return ES_

def _verify_confidence_intervals(Y_scens_, M_, S2_, z = 2.):
    
    N_in = 0.
    for i_sample in range(M_.shape[0]):
        for i_zone in range(M_.shape[1]):
            for i_hour in range(M_.shape[2]):
                y_    = Y_scens_[i_sample, i_zone, i_hour, :]
                mu    = M_[i_sample, i_zone, i_hour]
                std   = np.sqrt(S2_[i_sample, i_zone, i_hour])
                idx_  = (y_ > (mu - z*std)) & (y_ < (mu + z*std))
                N_in += idx_.sum()
            
    return N_in/(Y_scens_.shape[0]*Y_scens_.shape[1]*Y_scens_.shape[2]*Y_scens_.shape[3])

Y_     = _model['testing_targets']
Y_hat_ = _model['samples']



M_  = _model['mean']
S2_ = _model['standard_deviation']

print(_verify_confidence_intervals(Y_hat_, M_, S2_, z = 1.96))
print(_ES(Y_, Y_hat_))
print(_VS(Y_, Y_hat_))
```

```python
path_to_mdls = r'/Users/Guille/Desktop/caiso_power/models/'

resources_ = ['load', 'solar', 'wind']
resources_ = ['wind']

sl_models_ = ['OMP', 'lasso', 'elastic_net', 'group_lasso']
dl_models_ = ['GPR']
scores_    = ['ES', 'IS', 'VS']

z = 1.96
alpha = 0.05

z = 1.645
alpha = 0.1

def _interval_score(y_, alpha, lower, upper):
    score  = (upper - lower) 
    score += (2./alpha)*(lower - y_) * (1*(y_ < lower))
    score += (2./alpha)*(y_ - upper) * (1*(y_ > upper))
    return score

def interval_score(
    observations,
    alpha,
    q_dict=None,
    q_left=None,
    q_right=None,
    percent=False,
    check_consistency=True,
):
    """
    Compute interval scores (1) for an array of observations and predicted intervals.
    
    Either a dictionary with the respective (alpha/2) and (1-(alpha/2)) quantiles via q_dict needs to be
    specified or the quantiles need to be specified via q_left and q_right.
    
    Parameters
    ----------
    observations : array_like
        Ground truth observations.
    alpha : numeric
        Alpha level for (1-alpha) interval.
    q_dict : dict, optional
        Dictionary with predicted quantiles for all instances in `observations`.
    q_left : array_like, optional
        Predicted (alpha/2)-quantiles for all instances in `observations`.
    q_right : array_like, optional
        Predicted (1-(alpha/2))-quantiles for all instances in `observations`.
    percent: bool, optional
        If `True`, score is scaled by absolute value of observations to yield a percentage error. Default is `False`.
    check_consistency: bool, optional
        If `True`, quantiles in `q_dict` are checked for consistency. Default is `True`.
        
    Returns
    -------
    total : array_like
        Total interval scores.
    sharpness : array_like
        Sharpness component of interval scores.
    calibration : array_like
        Calibration component of interval scores.
        
    (1) Gneiting, T. and A. E. Raftery (2007). Strictly proper scoring rules, prediction, and estimation. Journal of the American Statistical Association 102(477), 359378.    
    """

    if q_dict is None:
        if q_left is None or q_right is None:
            raise ValueError(
                "Either quantile dictionary or left and right quantile must be supplied."
            )
    else:
        if q_left is not None or q_right is not None:
            raise ValueError(
                "Either quantile dictionary OR left and right quantile must be supplied, not both."
            )
        q_left = q_dict.get(alpha / 2)
        if q_left is None:
            raise ValueError(f"Quantile dictionary does not include {alpha/2}-quantile")

        q_right = q_dict.get(1 - (alpha / 2))
        if q_right is None:
            raise ValueError(
                f"Quantile dictionary does not include {1-(alpha/2)}-quantile"
            )

    if check_consistency and np.any(q_left > q_right):
        raise ValueError("Left quantile must be smaller than right quantile.")

    sharpness = q_right - q_left
    calibration = (
        (
            np.clip(q_left - observations, a_min=0, a_max=None)
            + np.clip(observations - q_right, a_min=0, a_max=None)
        )
        * 2
        / alpha
    )
    if percent:
        sharpness = sharpness / np.abs(observations)
        calibration = calibration / np.abs(observations)
    total = sharpness + calibration
    return total, sharpness, calibration

_results = {}
for resource in resources_:
    _results[resource] = {}
    
    for score in scores_:
        _results[resource][score] = {}

        for sl_model in sl_models_:
            _results[resource][score][sl_model] = {}
            
            for dl_model in dl_models_:
                filename = '{}-{}-{}-{}.pkl'.format(resource, sl_model, dl_model, score)
                print(filename)
                with open(path_to_mdls + filename, 'rb') as f:
                    _model = pickle.load(f)                    

                print(_model['bayesian_scoring'])

                _results[resource][score][sl_model][dl_model] = _model['bayesian_scoring_all'][score].to_numpy()
                
                for N_samples in [243]:
                    N_in = 0
                    IS = 0
                    SS = 0

                    for i_zone in range(_model['mean'].shape[1]):
                        for i_hour in range(_model['mean'].shape[2]):
                            y_   = _model['testing_targets'][:N_samples, i_zone, i_hour]
                            mu_  = _model['mean'][:N_samples, i_zone, i_hour] 
                            std_ = np.sqrt(_model['standard_deviation'][:N_samples, i_zone, i_hour])
                            lower = mu_ - z*std_
                            upper = mu_ + z*std_
                            idx_ = (y_ > lower) & (y_ < upper)
                              
                            N_in += idx_.sum()

                            SS += interval_score(y_, alpha, q_right = upper, q_left = lower)[1].sum()
                            IS += _interval_score(y_, alpha, lower, upper).sum()
                    N = N_samples*_model['mean'].shape[1]*_model['mean'].shape[2]
                    print(N_samples, i_zone, IS/N, SS/N, N_in/N)
```
